```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('./Datasets/stack-overflow-developer-survey-2023/survey_results_public.csv')
schema_df = pd.read_csv('./Datasets/stack-overflow-developer-survey-2023/survey_results_schema.csv')
```

```python
df.head()
```

```python
# To get the data of people who work Remote
df[df['RemoteWork']=='Remote']
```

```python
# masking with loc
df.loc[(df['RemoteWork']=='Remote'), 'EdLevel']
```

```python
# To deal with multiple conditions 
# and ==> &
# or ==> |
# ~ ==> Not
```

```python
# To get details of persons who work remote and have 5 years of coding experience
df[(df['RemoteWork'] == 'Remote') & (df['YearsCode']=="5")]
```

```python
high_salary = df['ConvertedCompYearly'] > 70000
```

```python
# get the country, Industry and Programing Languages of people earning high salary
df.loc[high_salary, ['Country', 'Industry', 'LanguageHaveWorkedWith']]
```

```python
# Filter the above data by countries
countries = ['United States of America', 'Canada', 'India']
countries_filter = df['Country'].isin(countries)
```

```python
df.loc[(high_salary & countries_filter), ['Country', 'Industry', 'LanguageHaveWorkedWith']]
```

```python
# further filter the above result on Languages
python_filter = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
```

```python
df.loc[(high_salary & countries_filter & python_filter), ['Country', 'Industry', 'LanguageHaveWorkedWith']]
```
