```python
# df[filter][col_name] = new_value (WON'T work)
# df.loc[filter, col_name] = new_value (WORKS)
```

```python
# Modifying multiple rows of data
df['REMOTEWORK'] = df['REMOTEWORK'].str.lower()
```

```python
df.head()
```

```python
# function to replace 'hybrid (some remote, some in-person)' with just 'Hybrid' 
def modify_hybrid(worktype):
    if isinstance(worktype, str) and worktype.strip()[:6]=='hybrid':
        return 'hybrid'
    else:
        return worktype
```

```python
df['REMOTEWORK'] = df['REMOTEWORK'].apply(modify_hybrid)
```
